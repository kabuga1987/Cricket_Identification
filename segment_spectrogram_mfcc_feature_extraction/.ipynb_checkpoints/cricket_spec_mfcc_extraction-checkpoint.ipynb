```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import librosa
import noisereduce as nr
from tqdm import tqdm
from scipy import signal
import skimage
```

```python
class Gibbon_spectrogram_extraction:

    def __init__(self, path_audio, path_anno, path_save_specs, cutoff, nyq_fr, new_sr):
        self.path_audio = path_audio
        self.path_anno = path_anno
        self.path_save_specs = path_save_specs
        self.cutoff = cutoff
        self.nyq_fr = nyq_fr
        self.new_sr = new_sr

    def read_audio_file(self, audio_file):
        signal, original_sr = librosa.load(self.path_audio+audio_file, sr = None)
        return signal, original_sr

    def reduce_noise(self, signal, original_sr):
        # source: https://github.com/timsainb/noisereduce
        reduced_noise = nr.reduce_noise(y = signal, sr = original_sr)
        return reduced_noise

    def butter_lowpass(self, cutoff, nyq_freq, order=4):
        normal_cutoff = float(cutoff) / nyq_freq
        b, a = signal.butter(order, normal_cutoff, btype='lowpass')
        return b, a

    def butter_lowpass_filter(self, data, cutoff_freq, nyq_freq, order=4):
        # Source: https://github.com/guillaume-chevalier/filtering-stft-and-laplace-transform
        b, a = self.butter_lowpass(cutoff_freq, nyq_freq, order=order)
        y = signal.filtfilt(b, a, data)
        return y

    def downsample_signal(self, signal, original_sr, new_sr):
        downsampled_signal = librosa.resample(signal, original_sr, new_sr, res_type='kaiser_fast')
        return downsampled_signal
    
    def preprocess_audio_file(self, audio_file):
        signal, original_sr = self.read_audio_file(audio_file)
        #reduced_noise_signal = self.reduce_noise(signal, original_sr)
        filtered_signal = self.butter_lowpass_filter(signal, self.cutoff, self.nyq_fr)
        downsampled_signal = self.downsample_signal(filtered_signal, original_sr, self.new_sr)
        #normalized_signal = downsampled_signal/np.abs(downsampled_signal).max()
        return downsampled_signal
        
    def read_segment_data(self):
        df =  pd.read_csv(self.path_anno)
        # slect "D" group only
        d = df[df.Group=="D"]
        audio_segment_st_end = list((audio,segment,start,end) for segment,start,end,audio,_ in d.to_records(index=False))
        return audio_segment_st_end

    def audio2_segment_indexes(self):
        
        audio_segment_st_end = self.read_segment_data()
        audio2data = {}
        
        for i, data in enumerate(audio_segment_st_end):
            audio, segment, st, end = data
            if audio not in audio2data: audio2data[audio] = [(segment,st,end)]
            else: audio2data[audio].append((segment,st,end))
        return audio2data

    def padding(self, signal,max = 10):
        
        length = len(signal)
        if length <= self.new_sr*max:
            dx = self.new_sr*max - length
            pad_left = dx//2
            pad_right = dx-pad_left
            padded = np.pad(signal,(pad_left, pad_right), mode = "constant", constant_values=0)
        else:
            dx =  length-self.new_sr*max
            st = int(dx*0.6)
            ed = dx-st
            padded = signal[st:-ed]
        return padded

    def audio_frame_to_mel_spectrogram(self, audio_frame, **kwargs):
        mel_spectrogram = librosa.feature.melspectrogram(y = audio_frame.astype(np.float32),**kwargs)
        return mel_spectrogram

    def scale_minmax(self, X, min=0.0, max=1.0):
        X_std = (X - X.min()) / (X.max() - X.min())
        X_scaled = X_std * (max - min) + min
        return X_scaled

    def segment_mfccs(self, segment_sig):
        mfccs = librosa.feature.mfcc(y = segment_sig,sr = self.new_sr,n_mfcc=20,n_mels = 128,fmin = 1000,fmax = 2000,
                                     n_fft = 1024,hop_length = 256)
        means = mfccs.mean(axis = 1)
        stdvs = mfccs.std(axis = 1)
        total = np.hstack([means, stdvs])
        return total
        

    
    # def Execution(self, **kwargs):
        
    #     odio2data = self.audio2_segment_indexes()

    #     for audio, data in tqdm(odio2data.items()):
    #         #os.makedirs(self.PATH_save_specs+audio[:-4]+"/")
    #         normalized_signal = self.preprocess_audio_file(audio + ".wav")
    #         for segment_data in data:
    #             name, st, end = segment_data
    #             segment = normalized_signal[self.newsr*st:self.new_sr*end]
    #             padded = self.padding(segment)
    #             # for saving original spectrograms
    #             mel_spec = self.audio_frame_to_mel_spectrogram(padded,**kwargs)
    #             image = librosa.core.power_to_db(mel_spec)
    #             #image_np = np.asmatrix(image)
    #             np.save(self.path_save_spec_arrays + name + ".npy", image)
                
    def Execution(self, **kwargs):
        
        odio2data = self.audio2_segment_indexes()
        
        segment_2_mfccs = {}

        for audio, data in tqdm(odio2data.items()):
            #os.makedirs(self.path_save_specs+audio+"/")
            normalized_signal = self.preprocess_audio_file(audio + ".wav")
            for segment_data in data:
                name, st, end = segment_data
                segment = normalized_signal[st:end]
                #padded = self.padding(segment)
                # for saving original spectrograms
                mel_spec = self.audio_frame_to_mel_spectrogram(segment,**kwargs)
                # scaled = self.scale_minmax(mel_spec , 0, 255).astype(np.uint8)
                # skimage.io.imsave(self.path_save_specs+audio+"/"+name+".png", scaled)
                # np.save("1_s_spectrograms/"+name+".npy", mel_spec)
                #image = librosa.core.power_to_db(mel_spec)
                #image_np = np.asmatrix(image)
                # Extract mfccs
                mfccs = self.segment_mfccs(segment)
                segment_2_mfccs[name] = mfccs
                
        mfcc_names = np.array(["x"+str(i) for i in range(40)])
        df = pd.DataFrame.from_dict(segment_2_mfccs, orient = 'index',columns = mfcc_names )
        df = df.rename_axis('Segment')
        df.to_csv("1_s_mfccs/D_mfccs.csv", index = True)
       
                
```

```python
p_odio = "bcd_audio_files/"
#p_anno = "Annotations_inffifteen.csv"
p_anno = "1_s_segments.csv"
p_arra = "1_s_images/"
cutoff = 2000
nyq_fr = 2400
new_sr = 4800
```

```python
WINDOW_LENGTH= 1024 # size of the window when applying STFT
HOPE_LENGTH = 256   # number of samples to jump between two consecutive windows
NUMBER_MELS = 128  # mels bins, this will be the HEIGHT of the mel spectrogram
SAMPLE_RATE = new_sr # sample rate used to down-sample the audio file
POWER = 1            # power of order 1, order i.e. squared energy
F_MIN = 1000        # minimum frequency to be detected
F_MAX  = 2000        # maximum frequency to be detected
```

```python
PARAMETERS = {"n_fft":WINDOW_LENGTH,
              "hop_length":HOPE_LENGTH,
              "n_mels":NUMBER_MELS,
              "sr":SAMPLE_RATE,
              "power":POWER,
              "fmin":F_MIN,
              "fmax":F_MAX
             }
```

```python
gibbon = Gibbon_spectrogram_extraction(p_odio, p_anno, p_arra, cutoff, nyq_fr, new_sr)
```

```python
gibbon.Execution(**PARAMETERS)
```
